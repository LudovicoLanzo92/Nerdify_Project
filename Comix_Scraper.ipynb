```python
def parse_details(fumetto):
    titolo = ""
    url = ""
    data_uscita = ""
    prezzo_pieno = ""
    prezzo_scontato = ""
    product_id = ""
    
    try:
        titolo = fumetto.find_elements_by_css_selector("div.col-md-9 > h3 > a")[0].text
        url = fumetto.find_elements_by_css_selector("div.col-md-9 > h3 > a")[0].get_attribute("href")
        data_uscita = fumetto.find_elements_by_css_selector("div.pnn-attributes > h4")[0].text
        prezzo_pieno = fumetto.find_elements_by_css_selector("div.col-md-9 > div > div:nth-child(1) > div.product-desc > div.price-box > p.old-price")[0].text
        prezzo_scontato = fumetto.find_elements_by_css_selector("div.col-md-9 > div > div:nth-child(1) > div.product-desc > div.price-box > p.special-price")[0].text
        product_id = fumetto.find_elements_by_css_selector("a")[0].get_attribute("href").split('/')[-1].split('-')[0]
    except:
        pass
    return{'titolo' : titolo,
                              'url' : url,
                              'data_uscita' : data_uscita,
                              'prezzo_pieno' : prezzo_pieno,
                              'prezzo_scontato' : prezzo_scontato,
                              'product_id' : product_id}
        
    
pprint.pprint(parse_details(fumetto))
```

```python
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
details = []

for product_id, fumetto in tqdm(df_dettaglio_fumetto.iterrows(), total=df_dettaglio_fumetto.shape[0]):
  

      
      prod_id = fumetto['product_id']
      link = fumetto["url"]  
      title = ''
      n_reprint = ''
      price_full = ''
      price_actual = ''
      release_date = ''
      publisher = ''
      author = ''
      pages = ''
      size = ''
      contents = ''
      binding = ''
      colors = ''
      isbn = ''
      description = ''
      

      try:
        wd.set_window_size(1920, 1080)
        wd.get(link)
        try:
          WebDriverWait(wd, 1)
        except Exception as e:
          continue

        #wd.save_screenshot(f'screenshot_{product_id}.png')
        time.sleep(2)
        title_list = wd.find_elements_by_css_selector("div>h1")
        n_reprint_list = wd.find_elements_by_css_selector("p.reprint")
        price_full_list = wd.find_elements_by_css_selector("p.old-price>span.price")
        price_actual_list = wd.find_elements_by_css_selector("p.special-price")
        release_date_list = wd.find_elements_by_css_selector("#data-pubblicazione > h3")
        publisher_list = wd.find_elements_by_css_selector("#linea-editoriale > h3")
        author_list = wd.find_elements_by_css_selector("#authors > div")
        pages_list = wd.find_elements_by_css_selector("#pages > div")
        size_list = wd.find_elements_by_css_selector("#format > div")
        contents_list = wd.find_elements_by_css_selector("#includes > div")
        binding_list = wd.find_elements_by_css_selector("#includes > div")
        colors_list = wd.find_elements_by_css_selector("#includes > div")
        isbn_list = wd.find_elements_by_css_selector("#isbn > div")
        description_list = wd.find_elements_by_css_selector("div.col-md-12>div.product")
        

        if(len(title_list) > 0):
          title = title_list[0].text     
        if(len(n_reprint_list) > 0):
          n_reprint = n_reprint_list[0].text
        if(len(price_full_list) > 0):
          price_full = price_full_list[0].text
        if(len(price_actual_list) > 0):
          price_actual = price_actual_list[0].text
        if(len(release_date_list) > 0):
          release_date = release_date_list[0].text     
        if(len(publisher_list) > 0):
          publisher = publisher_list[0].text
        if(len(author_list) > 0):
          author = author_list[0].text        
        if(len(pages_list)>0):
          pages = pages_list[0].text
        if(len(size_list) > 0):
          size = size_list[0].text
        if(len(contents_list) > 0):
          contents = contents_list[0].text
        if(len(binding_list) > 0):
          binding = binding_list[1].text
        if(len(colors_list) > 0):
          colors = colors_list[2].text 
        if(len(isbn_list) > 0):
          isbn = isbn_list[0].text         
        if(len(description_list) > 0):
          description = description_list[0].text        
          
      except Exception as e:
        print(e)
      details.append({'url': link,
                      'title': title,
                      'n_reprint': n_reprint,
                      'price_full': price_full,
                      'price_actual': price_actual,
                      'release_date': release_date,
                      'publisher': publisher,
                      'author': author,
                      'pages': pages,
                      'size': size,
                      'contents': contents,
                      'binding': binding,
                      'colors': colors,
                      'isbn': isbn,                      
                      'description': description,
                      'product_id' : prod_id})
  
print('numero righe lette ' + str(len(details)))
pprint.pprint(details)
```
